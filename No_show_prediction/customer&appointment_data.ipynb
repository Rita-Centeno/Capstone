```python
# Importing the necessary libraries
import pandas as pd
import random as rd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Defining the paths and opening the data
path = "C:\\Users\\ritaf\\Documents\\ESTUDOS\\NOVA IMS\\3ยบ Ano\\Capstone Project\\Projeto\\data\\"
appointment = pd.read_csv(path + 'no_show.csv')
customer = pd.read_csv(path + 'customer_data.csv')
hdi = pd.read_excel(path + 'HDI.xls')
```

---------

#### Dealing with patient and appointment data

```python
# Renaming column
appointment.rename(columns={'CountryofOriginHDI (Year-1)': 'CountryofOriginHDI'}, inplace=True)
```

```python
# Selecting 150 random appointment observatios to merge with the customer data
appointment = appointment.sample(n=150, replace=False, random_state=42)
```

```python
# Selecting ony the desired columns from the data
features = ['AppointmentID', 'AppointmentWeekNumber', 'AppointmentDayOfMonth', 'AppointmentHour', 'WeekendConsults', 'WeekdayConsults', 'Adults', 'Children', 'Babies', 'AffiliatedPatient', 'PreviousAppointments', 'PreviousNoShows', 'LastMinutesLate', 'OnlineBooking', 'AppointmentChanges', 'BookingToConsultDays', 'ParkingSpaceBooked', 'SpecialRequests', 'NoInsurance', 'ExtraExamsPerConsult', 'DoctorAssigned', 'ConsultPriceEuros', '%PaidinAdvance', 'CountryofOriginHDI', 'NoShow']
appointment = appointment.loc[:, features]
appointment.head()
```

```python
# Reseting the index to merge the appointment and customer data
appointment.reset_index(drop=True, inplace=True)
```

```python
# Merging the appointment and customer data
complete = pd.concat([appointment, customer], axis=1)
complete
```

```python
# Checking the whole set of columns
complete.columns
```

```python
# Dividing the data into patients and appointments data
patients = complete[['Adults', 'AffiliatedPatient', 'Age', 'Babies', 'Children', 'CountryofOriginHDI', 'Email', 'ExtraExamsPerConsult', 'FirstName', 'LastMinutesLate', 'LastName', 'NoInsurance', 'PreviousAppointments', 'PreviousNoShows', 'Username', 'WeekdayConsults', 'WeekendConsults']]
appointments = complete[['%PaidinAdvance', 'AppointmentChanges', 'AppointmentDayOfMonth', 'AppointmentHour', 'AppointmentWeekNumber', 'BookingToConsultDays', 'ConsultPriceEuros', 'DoctorAssigned', 'NoShow', 'OnlineBooking', 'ParkingSpaceBooked', 'SpecialRequests', 'Username', 'AppointmentID']]
```

```python
# Change the ordering of the appointment dataset rows and deleting the appointment ID column
appointments.sort_values(by=['AppointmentID'], inplace=True)
appointments.reset_index(drop=True, inplace=True)
appointments
```

```python
# Export the DataFrame to a CSV file
appointments.to_csv(path + 'appointments.csv', index=True)
patients.to_csv(path + 'patients.csv', index=True)
```
